```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'vgsales.csv'
data = pd.read_csv(file_path)

# Total Global Sales by Genre
genre_sales = data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# Total Global Sales by Platform
platform_sales = data.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

# Yearly Global Sales Trend
yearly_sales = data.groupby('Year')['Global_Sales'].sum().sort_values(ascending=True)

# Top 10 games by global sales
top_10_games = data.nlargest(10, 'Global_Sales')[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].set_index('Name')

# Plotting the graphs
plt.figure(figsize=(20, 20))

# Plot 1: Total Global Sales by Genre
plt.subplot(2, 2, 1)
genre_sales.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')

# Plot 2: Total Global Sales by Platform
plt.subplot(2, 2, 2)
platform_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')

# Plot 3: Yearly Global Sales Trend
plt.subplot(2, 2, 3)
yearly_sales.plot(kind='line', marker='o', color='salmon')
plt.title('Yearly Global Sales Trend')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')

# Plot 4: Regional Sales Distribution for Top 10 Games
top_10_games.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.title('Regional Sales Distribution for Top 10 Games')
plt.xlabel('Game Name')
plt.ylabel('Sales (in millions)')
plt.legend(title='Region')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('vgsales.csv')

# Total Global Sales by Genre
genre_sales = data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# Plotting the graph
plt.figure(figsize=(10, 6))
genre_sales.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Global Sales by Platform
platform_sales = data.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

# Plotting the graph
plt.figure(figsize=(10, 6))
platform_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Yearly Global Sales Trend
yearly_sales = data.groupby('Year')['Global_Sales'].sum().sort_values(ascending=True)

# Plotting the graph
plt.figure(figsize=(10, 6))
yearly_sales.plot(kind='line', marker='o', color='salmon')
plt.title('Yearly Global Sales Trend')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()

```

```python
# Top 10 games by global sales
top_10_games = data.nlargest(10, 'Global_Sales')[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].set_index('Name')

# Plotting the graph
plt.figure(figsize=(12, 8))
top_10_games.plot(kind='bar', stacked=True, colormap='tab20')
plt.title('Regional Sales Distribution for Top 10 Games')
plt.xlabel('Game Name')
plt.ylabel('Sales (in millions)')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
corr_matrix = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Sales')
plt.show()
```

```python
# Aggregating sales by genre and year
genre_year_sales = data.groupby(['Year', 'Genre'])['Global_Sales'].sum().unstack().fillna(0)

# Plotting the sales by genre over time
plt.figure(figsize=(14, 8))
genre_year_sales.plot(kind='line', stacked=True, colormap='tab20', linewidth=2)
plt.title('Global Sales by Genre Over Time')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.legend(title='Genre')
plt.show()
```
